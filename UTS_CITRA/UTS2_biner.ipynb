***Image Resulotion***

```python
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]

print(f"Resolution: {img_height} x {img_width}")
print(f"Channels: {img_channel}")
```

***Convert Grayscale***

```python

gray_img = np.zeros_like(img)

for y in range(img_height):
    for x in range(img_width):
        red = img[y, x, 0]
        green = img[y, x, 1]
        blue = img[y, x, 2]
        alpha = img[y, x, 3]  # salin alpha
        gray = int((int(red) + int(green) + int(blue)) / 3)
        gray_img[y, x] = (gray, gray, gray, alpha)
```

```python
# View Grayscale Image
plt.imshow(gray_img)
plt.title("Image After Grayscale")
plt.axis("off")
plt.show()
```
