***Import*** ***Library***

```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

***Import Image***

```python
img = imageio.imread('/content/drive/My Drive/percobaan.jpg')
print(img)
```

***View Image Before***

```python
plt.imshow(img)
plt.title("Image Before")
plt.show()
```

***Image Resulotion***

```python
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]

print(img_height,'x', img_width)
print(img_channel)
```

***Convert Grayscale***

```python
for y in range(0, img_height):
    for x in range(0, img_width):
        red = img[y][x][0]
        green = img[y][x][1]
        blue = img[y][x][2]
        gray = (int(red) + int(green) + int(blue)) / 3
        img[y][x] = (gray, gray, gray)
```

***View After Image***

```python
plt.imshow(img)
plt.title("Image After")
plt.show()
```
